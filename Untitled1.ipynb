```python
import torch
from models.unet import UNET
# from utils import *
import torch
from models.unet import UNET
# from utils import *
from tqdm import tqdm
import matplotlib.pyplot as plt
from torchvision import transforms
import data
from data import relis3d
import cv2
device = 'cuda:0'
from PIL import Image
import data
from torchvision import transforms
```

```python
MODEL_PATH='./checkpoints/SourceOnlyTrain_rellis3d/epoch_0099'
target_loader = data.setup_loaders('semantickitti', './dataset/SemanticKitti/vallist.txt', 1)
targetloader_iter = enumerate(target_loader)
```

```python
_, batch = targetloader_iter.__next__()
```

```python
img, mask= batch
X, y = img.to(device), mask.to(device)
```

```python
net = UNET(in_channels=3, classes=10).to(device)
checkpoint = torch.load(MODEL_PATH)
net.load_state_dict(checkpoint['model_state_dict'])
```

```python
predictions = net(X)
predictions = torch.nn.functional.softmax(predictions, dim=1)
pred_labels = torch.argmax(predictions, dim=1) 
pred = pred_labels.to('cpu')
```

```python
SMOOTH = 1e-6
outputs = pred
labels = mask
intersection = (outputs & labels).float().sum((1, 2)) 
union = (outputs | labels).float().sum((1, 2)) 
iou = (intersection + SMOOTH) / (union + SMOOTH)
print(iou)
```
